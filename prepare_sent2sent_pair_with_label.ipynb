```python
import pandas as pd
from tqdm import tqdm
import nltk
from utils import load_json, dump_json
```

```python
ps = pd.read_csv("PersonalStatements_16Nov.csv",encoding='unicode_escape')
```

```python
dic ={"sent1":[],"sent2":[],'label':[]}
sent_pairs = []
p = []
for i, raw in ps.iterrows():
    p+=raw["ps_content"].split("\n")
```

```python
len(p)
```

```python
sentences = []
for para in p:
    sents = nltk.sent_tokenize(para)
    sentences.extend(sents)
from sklearn.utils import shuffle
sentences = shuffle(shuffle(sentences))
for i in tqdm(range(len(sentences)-9)):
    dic00 ={"sent1": sentences[i],"sent2":sentences[i+2],'label':0}
    dic01 ={"sent1": sentences[i],"sent2":sentences[i+5],'label':0}
    dic02 ={"sent1": sentences[i],"sent2":sentences[i+7],'label':0}
    sent_pairs.append(dic00)
    sent_pairs.append(dic01)
    sent_pairs.append(dic02)
    
```

```python
for para in p:
    sents = nltk.sent_tokenize(para)
    for i in tqdm(range(len(sents)-1)):
        dic1 ={"sent1": sents[i],"sent2":sents[i+1],'label':1}
#         dic0 ={"sent1": sents[i+1],"sent2":sents[i],'label':0}
        sent_pairs.append(dic1)
#         sent_pairs.append(dic0)
```

```python
sent_pairs = [sents for sents in sent_pairs if len(sents["sent1"])>30 and len(sents["sent2"])> 30]
```
